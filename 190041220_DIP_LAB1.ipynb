```python
def histogram(image, title):
  plt.hist(image.ravel(), bins = 256, range = [0, 256], histtype = 'step')
  plt.title(title)
  plt.show()
```

```python
gray_img = plt.imread("gray.tif")
histogram(gray_img, "Histogram")
```

Define hist function by your own:

```python
def histogram_manual(image):
    intensity_values = np.arange(256)
    pixel_counts = np.zeros(256, dtype=int)

    m, n = image.shape

    for i in range(m):
        for j in range(n):
            intensity = image[i, j]
            pixel_counts[intensity] += 1

    return intensity_values, pixel_counts


```

```python
img = cv2.imread('new.jpg', 0)
intensity_values, pixel_counts = histogram_manual(img)

plt.stem(intensity_values, pixel_counts)
plt.xlabel('Intensity')
plt.ylabel('Pixel Count')
plt.title('Manual Histogram')

plt.show()
```
